#### 7. What's the 1-gram language model;

Ans:把统计语言模型就是计算一个句子出现概率大小的模型，常用的N元语法模型，就是将一个句子中的词看成一个有序序列，依次计算词的概率，并计算句子的概率值，当n=1时，称为1-gram language model，即所有的词的出现和其他词是否出现没有关系。此操作主要是为了简化计算。

统计语言模型：

$$P(w_1, w_2, \dots , w_n) = p(w_1)  p(w_2 | w_1)  p(w_3 | w_1, w_ 2) \dots p(w_n | w_1, w_2, \dots , w_{n-1}) = \prod_i^n p(w_i| w_1, w_2, \dots , w_{i-1})$$

1-gram language model:

$$P(w_1, w_2, \dots , w_n) = p(w_1)  p(w_2)  p(w_3 ) \dots p(w_n ) = \prod_i^n p(w_i) = \prod_i^n { \frac {C_{w_i}}{\sum_{\forall k} C_{w_k}}} = \prod_i^n {\frac {C_{w_i}}{N}}$$


#### 8. What's the disadvantages and advantages of 1-gram language model;

Ans:1-gram language model没有考虑到词之间的关系，很多词是伴随着上一个词，或下一个词，才出现的。如果只把词序列当作独立进行计算，就必然损失了词之间的关联性。因此需要解决这个问题出现了2-gram language model , n-gram language model.

#### 9. What't the 2-gram models;

Ans:2-gram models，只考虑前一个词出现的条件概率。

$$P(w_1, w_2, \dots , w_n) = p(w_1)  p(w_2 | w_1)  p(w_3 | w_ 2) \dots p(w_n | w_{n-1}) = \prod_i^n p(w_i| w_{i-1}) $$

$$p(w_i| w_{i-1}) = \frac {p(w_i, w_{i-1})}{p(w_{i-1})} = \frac {C_{w_i, w_{i-1}}}{C_{w_{i-1}}}$$

````python
#封装函数，得到语料库中独立单词频率字典和连续两个词的频率字典
def get_1_2_gram_word_freq_dict(FILE_CORPUS: str):
    '''
    根据语料库对每个单词行进独立统计，用于建立针对于每个单词的词典。
    FILE_CORPUS ：得到的纯文本字符串
    return： 针对每个单词频率的词典, 针对两个连续单词、频率的词典，单个词频率词典的最低频率值，2个连续词频率词典的最低频率值。
    '''
   
    WORD_TOKENS = list(jieba.cut(FILE_CORPUS)) #根据每个独立的单词，生成一个list
    word_freq_dict = Counter(WORD_TOKENS)
    #用于数据平滑
    word_low_freq = word_freq_dict.most_common()[-1][-1]

    _2_WORD_TOKENS = [WORD_TOKENS[i] + WORD_TOKENS[i+1] for i in range(len(WORD_TOKENS) - 1)]
    _2_word_freq_dict = Counter(_2_WORD_TOKENS)
    #用于数据平滑
    _2_word_low_freq = word_freq_dict.most_common()[-1][-1]
        
    return word_freq_dict, _2_word_freq_dict, word_low_freq, _2_word_low_freq
    
````

````python
word_freq_dict, _2_word_freq_dict, word_low_freq, _2_word_low_freq = get_1_2_gram_word_freq_dict(FILE_CORPUS_MOVIE_COMMENTS_DATA)
````

````python
def get_1_word_frequence(word, word_dict = word_freq_dict, word_low_freq = word_low_freq):
    '''
    得到一个单词在独立词频率字典中的频率数
    '''
    word_freq = 0
    if word in word_dict.keys():
        word_freq = word_dict[word]
    else:
        word_freq = word_low_freq
    
    return word_freq
             
````

````python
def get_2_word_frequence(word, word_dict =_2_word_freq_dict, word_low_freq = _2_word_low_freq):
    '''
    得到连续2个单词在2个连续词频率字典中的频率数
    '''
    word_freq = 0
    if word in word_dict.keys():
        word_freq = word_dict[word]
    else:
        word_freq = word_low_freq
    
    return word_freq
        
````

````python
def get_probablity_by_1_gram(sentence : str, word_dict = word_freq_dict):
    
    sentence_pro = 1
    
    if not sentence :
        raise("error: sentence is None")
    word_cut = list(jieba.cut(sentence))
    word_all_count = len(word_dict)
    for i in range(len(word_cut)):
        #print (word_cut[i])
        #print('get_1_word_frequence(word_cut[i]): ', get_1_word_frequence(word_cut[i]))
        #print('word_all_count: ', word_all_count)
        sentence_pro *= get_1_word_frequence(word_cut[i]) / word_all_count
    return sentence_pro
            
````

````python
def get_probablity_by_2_gram(sentence : str, word_dict = word_freq_dict, _2_word_dict = _2_word_freq_dict):
    
    if not sentence :
        raise("error: sentence is None")
    word_cut = list(jieba.cut(sentence))
    #得到句子第一个元素的概率
    sentence_pro = get_1_word_frequence(word_cut[0]) / len(word_dict)
    
    for i in range(len(word_cut) - 1 ):
        #print (word_cut[i]+ word_cut [i+1])
        #print('get_2_word_frequence(word_cut[i] + word_cut [i+1] ): ', get_2_word_frequence(word_cut[i] + word_cut [i+1] )) 
        #print('get_1_word_frequence(word_cut[i]) : ', get_1_word_frequence(word_cut[i]))
        sentence_pro *= get_2_word_frequence(word_cut[i] + word_cut [i+1] ) / get_1_word_frequence(word_cut[i])
    
    
    return sentence_pro
````

````python
_1_gram_value = get_probablity_by_1_gram('我是个聪明的女子')
print(_1_gram_value)
````

````python
def generate_best(input_rules, num_sentence, ): # you code here
    gen_n_2 = generate_n(n = num_sentence , rules = input_rules)
    pro_sentece_list = [(get_probablity_by_2_gram(i_gen), i_gen) for i_gen in gen_n_2]
    
    decrease_pro_sentece_list = sorted(pro_sentece_list, key=lambda x: x[0], reverse = True) #按照概率的大小进行排序，从大到小
    
    [print(i_sentence) for i_sentence in decrease_pro_sentece_list]
    
    return decrease_pro_sentece_list[0][1]
    
````
